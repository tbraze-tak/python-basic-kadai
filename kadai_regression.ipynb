```python
########### 重回帰 ##########

import statsmodels.api as sm
import pandas as pd
import seaborn as sns
import numpy as np

# データの読み込み
data = pd.read_csv(r"/content/california_housing_cleansing.csv")
#data.head()

data=data.drop(columns=["Unnamed: 0"])
#data.head()
#print(data.head())

#回帰分析に使用するデータの指定

### 説明変数の指定 ###
x = data.iloc[ :, [0,1,2,3,6] ]
x.head()
print(x.head())


### 目的変数の指定 ###
y = data.iloc[:, 5]
print(y.head())

#全要素が1の列を説明変数の先頭に追加
X = sm.add_constant(x)

#予測モデルの設定(インスタンス化)[最小二乗法]
model = sm.OLS(y, X)

#予測モデルの学習
results = model.fit()

#予測モデルの評価
print(results.summary())

# 新しいデータ
x_new = np.array([[1,8, 41, 500, 38, 2],
                  [1,5, 10, 1000, 40, 1],
                  [1,2.5, 5, 1700, 37, 2.7]])

#予測
predictions = results.predict(x_new)

print(predictions)

```

```python
########### 単回帰 ##########

import statsmodels.api as sm
import pandas as pd
import seaborn as sns
import numpy as np

# データの読み込み
data = pd.read_csv(r"/content/california_housing_cleansing.csv")
#data.head()

data=data.drop(columns=["Unnamed: 0"])
#data.head()
#print(data.head())

#回帰分析に使用するデータの指定

### 説明変数の指定 ###
x = data.iloc[ :, [0] ]
x.head()
print(x.head())


### 目的変数の指定 ###
y = data.iloc[:, 5]
print(y.head())

#全要素が1の列を説明変数の先頭に追加
X = sm.add_constant(x)

#予測モデルの設定(インスタンス化)[最小二乗法]
model = sm.OLS(y, X)

#予測モデルの学習
results = model.fit()

#予測モデルの評価
print(results.summary())

# 新しいデータ
x_new = np.array([[1,8],
                  [1,5],
                  [1,2.5]])

#予測
predictions = results.predict(x_new)

print(predictions)
```
